### Retrieving and Tidying Data

**In the next couple of cells, I will read in the data from three csv files: Pitching.csv, Salaries.csv, People.csv.

```python
# read in salaries.csv, pitching.csv, and people.csv into dataframes
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
salaries_df = pd.read_csv("./data/Salaries.csv")
display(salaries_df.head())
pitching_df = pd.read_csv("./data/Pitching.csv")
display(pitching_df.head())
people_df = pd.read_csv("./data/People.csv")
display(people_df.head())
```

### Comparing Pitcher Salary to Total Salary

**Use line graphs to compare average pitcher salary to average total player salary from 1985 to 2014**

```python
# trim salaries dataframe to relevant years
salaries_current = salaries_df[salaries_df["yearID"] >= 1985]
salaries_current.head()
# get mean player salary
mean_player_salaries = salaries_current.groupby("yearID")["salary"].mean()/1000
# get mean pitcher salaries
mean_pitcher_salaries = pitching_salaries.groupby("yearID")["salary"].mean()/1000
# plot mean salaries
mean_plot = mean_player_salaries.plot.line(legend=True)
mean_pitcher_salaries.plot.line(ax=mean_plot, legend=True)
mean_plot.set_title("Mean Salaries Over Time")
mean_plot.set_ylabel("Mean Salary (in thousands of dollars)")
mean_plot.set_xlabel("Year")
mean_plot.legend(["All Players", "Pitchers"])
```

**From this graph we can see that from 1985 to 2010 pitchers were paid more than the average baseball player, but in the years since 2010 there has been a drastic change as now pitchers are getting paid much less than the average player, most likely due to large contracts handed out to position players in the past decade.**

### Track Pitching Statistics Over Time 

**Use line graphs to track certain pitching statistics over time, including earned run average and strikeouts**

```python
# trim pitching dataframe to just past 100 years
pitching_trim = pitching[pitching["yearID"] > 1914]

# plot total strikeouts over time on line graph
total_strikeouts = pitching_trim.groupby("yearID")["SO"].sum()
so_plot = total_strikeouts.plot.line(title="Total Strikeouts Over Time") 
so_plot.set_ylabel("Total Strikeouts")
so_plot.set_xlabel("Year")
```

**From the graph we can see that since 1980 total strikeouts have gone up a lot but recently there has been a decrease**

```python
# plot earned run average over time on line graph
mean_era = pitching_trim.groupby("yearID")["ERA"].mean()
era_plot = mean_era.plot.line(title="Earned Run Average Over Time") 
era_plot.set_ylabel("ERA")
era_plot.set_xlabel("Year")
```

**From the graph we can see that ERA has fluctuated over the years, with it being the lowest from 1960 to the end of the century, which is when pitching was peaking**
